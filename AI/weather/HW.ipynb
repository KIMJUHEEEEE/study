```python
train_year = (res1["연"]<=2019)
test_year = (res1["연"]>=2020)
interval = 6
```

```python
def make_data(data):
    #학습데이터
    x = []
    #결과데이터
    y = []
    temps=list(data['기온'])
    
    for i in range(len(temps)):
        if i<interval: continue
        y.append(temps[i])
        xa=[]
        for p in range(interval):
            d=i+p-interval
            xa.append(temps[d])
        x.append(xa)
    
    return (x,y)
```

```python
#train, test data 준비
x_train,y_train = make_data(res1[train_year]) #2006_2015
x_test,y_test = make_data(res1[test_year]) #2016
```

```python
#Linear Regression Algorithm
model = LinearRegression(normalize=True)
#학습하기
model.fit(x_train,y_train)
#검증하기
y_predict = model.predict(x_test)
```

```python
#결과를 그래프로 표시
plt.figure(figsize=(16,10),dpi=100)
plt.plot(y_test,c='r')
plt.plot(y_predict,c='b')
plt.savefig('weather-temp.png')
plt.show()
```

```python
#오차 출력
print(y_predict-y_test)
diff_y = abs(y_predict-y_test)
print('평균=',sum(diff_y)/len(diff_y))
print('최대 오차=',max(diff_y))
```
